```python
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as sfa
```

```python
sm.formula.ols
```

```python
file=r"D:\Data Science By Sir\16.12.2019 One_Way_ANOVA\PlantGrowth.csv"
df=pd.read_csv(file)
```

```python
df.head()
```

```python
df.boxplot('weight',by='group')
```

```python
mod=sm.formula.ols(formula='weight ~ group', data=df).fit()
```

```python
aov_table=sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```
